```python
import matplotlib.pyplot as plt
import numpy as np
import collections

Fin = open("Points.txt", "r")

#plt.figure(figsize=(30,30)) plt.figure(figsize=(15,15))
plt.figure(figsize=(30,30))
ax = plt.axes()

N = int(Fin.readline())
M = int(Fin.readline())
Points = []
Xk = []
Yk = []
font = {'family': 'serif',
        'color':  'red',
        'weight': 'bold',
        'size': 5
}
for i in range(N):
    temp = list(map(float,Fin.readline().strip().replace('(', '').replace(')', '').replace(' ', '').split(',')))
    Points.append((temp[0], temp[1]))
    Xk.append(temp[0])
    Yk.append(temp[1])
    plt.scatter(temp[0], temp[1], color='m', s=300)
    plt.text(temp[0]+0.005, temp[1]-0.01, str(i), size=20, weight='bold')

#plt.scatter(Xk, Yk, color='k')
#plt.show()
#print(Points)
#Xk = [i[0] for i in Points]
#Yk = [i[1] for i in Points]

#ax.arrow(0.5, 0.1, 0.1, 0.5, head_width=0.007, head_length=0.01, fc='b', ec='g')
#ax.arrow(0.5, 0.1, 0.1, 0.5 , head_width=0.007, head_length=0.01, ec='g')

#plt.plot(Xk, Yk, 'ro')

Fin.readline()
Edges = collections.defaultdict(list)
for i in range(M):
    str_temp = Fin.readline().split()
    full_temp =  list(str_temp)
    temp = list(map(int, full_temp[0:2]))
    lengths = list(map(float, full_temp[2:4]))
    Edges[temp[0]].append((temp[1], lengths))
    #print(lengths)
    #print("Temp {}".format(temp))
    #plt.plot([Points[temp[0]][0], Points[temp[1]][0]], [Points[temp[0]][1], Points[temp[1]][1]], color='b')
    dx = Points[temp[1]][0] - Points[temp[0]][0]
    dy = Points[temp[1]][1] - Points[temp[0]][1]
    ax.arrow(Points[temp[0]][0], Points[temp[0]][1], dx, dy, head_width=0.007, head_length=0.02, fc='b', ec='b')
    #plt.text(Points[temp[0]][0]+dx/2, Points[temp[0]][1]+dy/2, str(lengths[1]), size=20, weight='bold')
    

Fin.close()

Fin2 = open("Shortest_Paths.txt", "r")
path = []
myline = Fin2.readline()
while myline:
    path.append(int(myline))
    myline = Fin2.readline()

print(path)
for i in range(len(path)-1):
    plt.plot([Points[path[i]][0], Points[path[i+1]][0]], [Points[path[i]][1], Points[path[i+1]][1]], color='g', linewidth=20, fillstyle='top')

Fin2.close()


Fin3 = open("Deviated_Edges.txt", "r")
temp =  list(Fin3.readline().split(","))
del temp[-1]
deviated = []
for tv in temp:
    tm = list(map(int,tv.split()))
    #print(tm)
    deviated.append((tm[0], tm[1]))
    
print(deviated)

for jv in deviated:
    plt.plot([Points[jv[0]][0], Points[jv[1]][0]], [Points[jv[0]][1], Points[jv[1]][1]], color='r', linewidth=4)

Fin3.close()



Fin4 = open("Robust_Paths.txt", "r")
colors = ['m', 'y', 'k', 'c']
myline = Fin4.readline()
j = 0
while myline:
    j = j+1
    print("myline: {}".format(myline.split()))
    path = list(map(int,myline.split()))
    myline = Fin4.readline()
    for i in range(len(path)-1):
        plt.plot([Points[path[i]][0], Points[path[i+1]][0]], [Points[path[i]][1], Points[path[i+1]][1]], color=colors[j-1], linewidth=10, fillstyle='top')
    
Fin4.close()

plt.show()
```
